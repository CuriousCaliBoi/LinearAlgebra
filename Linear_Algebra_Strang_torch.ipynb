```python
import numpy as np
import matplotlib.pyplot as plt

def shear_matrix(k):
    return np.array([[1, k], [0, 1]])

def apply_transformation(matrix, points):
    return np.dot(matrix, points)

# Create a grid of points
x = np.linspace(-10, 10, 20)
y = np.linspace(-10, 10, 20)
X, Y = np.meshgrid(x, y)
points = np.vstack([X.ravel(), Y.ravel()])

# Shear transformation matrix
k = 1.5  # Shear factor
shear_mat = shear_matrix(k)

# Apply shear transformation
transformed_points = apply_transformation(shear_mat, points)

# Plot original and transformed grid
plt.figure(figsize=(10, 5))

# Original grid
plt.subplot(1, 2, 1)
plt.plot(points[0], points[1], 'bo')
plt.title('Original Grid')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)

# Transformed grid
plt.subplot(1, 2, 2)
plt.plot(transformed_points[0], transformed_points[1], 'ro')
plt.title(f'Sheared Grid (k={k})')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)

plt.tight_layout()
plt.show()
```
